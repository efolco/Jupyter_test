### Quels sont les genres botaniques les plus représentés ? Quelle est la proportion par rapport au total ? Les objectifs de la Métro sont-ils respectés en terme de diversité ?

```python
especeCount = {} # dictionnaire comptant le nombre de chaque genre dans le tupple (genre, espece)
for key in genreEspece:
    if key[0] in especeCount.keys():
        especeCount[key[0]] += 1
    else:
        especeCount[key[0]] = 1
# print(especeCount)

# reverseEspeceCount = {j:i for i,j in especeCount.items()}
# print(max(reverseEspeceCount.keys()))
# reverseEspeceCount[max(reverseEspeceCount.keys())]

# max(especeCount.values())

nbByGenre = Counter([g for g, e in genreEspece])
totalGenre = sum(nbByGenre.values())

genrePlusRep = {}
# [k for k in nbByGenre if nbByGenre[key] ]
    
for key in nbByGenre:                              # pour clé dans le dico:
    value = nbByGenre[key]                         # je prends la première clé
    genrePlusRep[key] = 100*(value/totalGenre)     # pour la clé, je lui associe la valeur
                                                   #               100*la valeur pour la clé/total

orderedGenre = sorted(genrePlusRep.items(), key=lambda x: x[1], reverse=True)

nb = int(input("nb de genre ?"))
print("Les",nb,"genres botanique les plus représentés sont :\n")
for limit, i in enumerate(orderedGenre):
    print(f"{i[0]} : {i[1]:.2f}%")
    if limit == nb-1:
        break
if i[1] < 20:
        print("\nLes objectifs de la Métro sont respectés concernant les genres")
else:
    print("\nLes objectifs de la Métro ne sont pas respectés concernant les genres")

# pour comprendre :
# resultat = 42.42954446854
# print(f"le résultat est : {resultat:.2f} euro")
```
